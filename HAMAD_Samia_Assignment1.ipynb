```python
# (i)
# Using value_counts() function, display how many students have passed and how
# many have failed using following format where x stands for a number. The display will
# be given automatically by the function. The remark which has the highest number
# should come first).
# Pass x
# Fail x

df['Remark'] = df['FM'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')
df['Remark'].value_counts()
```

```python
# (j)
# Using groupby() function, display how many students have passed and how many
# have failed with their Sex in the following format where x stands for a number. For
# each group of Sex, it doesn’t matter whether Fail or Pass starts in the display.
df['Remark'] = df['FM'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')
grouped = df.groupby('Sex')['Remark'].value_counts()
print(grouped)
```

```python
# (k)
# Using groupby() function, display students failed and their grades for each Program

# First let grade the Final marks 
def assign_grade(fm):
    if 70 <= fm <= 100:
        return 'A'
    elif 60 <= fm < 70:
        return 'B+'
    elif 50 <= fm < 60:
        return 'B'
    elif 40 <= fm < 50:
        return 'C'
    elif 35 <= fm < 40:
        return 'D'
    else:
        return 'E'

df['Grade'] = df['FM'].apply(assign_grade)
```

```python
# state which marks is considerder as fail   
failedStudents = df[df['FM'] < 40]
```

```python
# grouped failed student by Program and grade score 
failed_grouped = failedStudents.groupby(['Program', 'Grade']).size()
```

```python
print(failed_grouped)
```

```python
# (l)
# Display the Fullname, Program, Final Mark and Grade for the first 3 best students in
# FM column. There is a number of possibilities. One of possibility is to use pandas
# nlargest function.
Beststudents = df.nlargest(3, 'FM')[['Fullname', 'Program', 'FM', 'Grade']]
print(Beststudents)
```

```python
# (m)
# Display the Fullname, Program, FM and Grade for the best student from each Program. That
# means pick one best student from each Program.

bestPerProgram_idx = df.groupby('Program')['FM'].idxmax()
bestStudents = df.loc[bestPerProgram_idx, ['Fullname', 'Program', 'FM', 'Grade']]
bestStudents
```

```python
# (o)
# Display the Fullname, Sex, Program, Grade, and Remark for Male students from either CEIT or
# CS who Failed. Also display how many records are there, in the format: There are … records
# which meets the criteria

male_students = df['Sex'] == 'M'
program_filter = df['Program'].isin(['CEIT', 'CS'])
fail_filter = df['Remark'] == 'Fail'
filtered = df[male_students & program_filter & fail_filter]
result = filtered[['Fullname', 'Sex', 'Program', 'Grade', 'Remark']]
print(result)
print(f"There are {len(filtered)} records which meet the criteria")
```
